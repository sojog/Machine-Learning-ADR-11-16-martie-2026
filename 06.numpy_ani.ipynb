```python
import numpy as np
```

```python
start_date = 1900
end_date = 2026


all_years = np.arange(start_date, end_date + 1)
all_years
```

```python
all_years [ all_years % 400 == 0 ]
```

```python
all_years [ (all_years % 4 == 0) & (all_years % 100 != 0) ]
```

```python
all_years [ (all_years % 400 == 0) | ((all_years % 4 == 0) & (all_years % 100 != 0)) ]
```

```python
divizibil_cu_400    = all_years % 400 == 0
divizibil_cu_4      = all_years % 4 == 0
nedivizibil_cu_100  = all_years % 100 != 0


all_years [ divizibil_cu_400 | (divizibil_cu_4 & nedivizibil_cu_100) ]
```

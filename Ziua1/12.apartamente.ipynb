```python
from scipy import stats
```

```python
result = stats.linregress(suprafete, preturi)
result
```

```python
result.slope, result.intercept
```

```python
plt.scatter(suprafete, preturi, c="red")

plt.plot(suprafete, 1700*suprafete)
plt.plot(suprafete, result.slope*suprafete + result.intercept, c="black")

plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
plt.scatter(suprafete, preturi, c="red")


plt.plot(suprafete, result.slope*suprafete + result.intercept, c="black")

plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
def f(x):
    return result.slope * x + result.intercept
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# mp, $$
apartamente = np.array([
   (30, 40_000), (39, 67_000), (45, 80_000), (55, 90_000), (60, 100_000), (65, 115_000), (70, 120_000), (75, 135_000), (80, 160_000)

])
suprafete = apartamente[:, 0]
preturi = apartamente[ : , 1]
```

```python
plt.scatter(suprafete, preturi, c="red")
plt.plot(suprafete, preturi)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

# pip install scipy

```python
from scipy import stats
```

```python
result = stats.linregress(suprafete, preturi)
result
```

```python
result.slope, result.intercept
```

```python
plt.scatter(suprafete, preturi, c="red")

plt.plot(suprafete, result.slope*suprafete + result.intercept, c="black")

plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
plt.scatter(suprafete, preturi, c="red")


plt.plot(suprafete, result.slope*suprafete + result.intercept, c="black")

plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
def f(x):
    return result.slope * x + result.intercept
```

```python
SUPRAFATA = 50
f(SUPRAFATA)
```

```python
plt.scatter(suprafete, preturi, c="red")
SUPRAFATA = 35
plt.scatter(SUPRAFATA, f(SUPRAFATA), c="green")
plt.plot(suprafete, result.slope*suprafete + result.intercept, c="black")

plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BLACK = 	np.array([0, 0, 0, 255])
YELLOW = np.array([252, 209, 22, 255])
RED = np.array([206, 17, 38, 255])
```

```python
np.zeros(10)
```

```python
np.ones(10)
```

```python
np.ones(10) * 5
```

```python
WIDTH = 100
HEIGHT = 150
#                        4 - RGBA
SHAPE_IMAGINE = (HEIGHT, WIDTH,  4)

imagine_alba = np.ones(SHAPE_IMAGINE, dtype=np.uint8) * 255
imagine_alba
```

```python
from PIL import Image
```

```python
edited_image = Image.fromarray(imagine_alba)
edited_image.save("imagine_alba_10_x_10.png")
```

```python
imagine_alba[ : , : WIDTH // 3] = BLACK
```

```python
imagine_alba[ : , WIDTH // 3 : 2 * WIDTH // 3 ] = YELLOW
```

```python
imagine_alba[ : , 2 * WIDTH // 3 :  ] = RED
```

```python
edited_image.save("steag_germania.png")
```

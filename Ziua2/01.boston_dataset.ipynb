```python
x = df.drop("medv", axis=1) # axis=0 orizontala si axis=1 este pe verticala
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model.fit(x_train, y_train)
```

```python
y_prezis = model.predict(x_train)
y_prezis
```

```python
real_vs_prezis_df = pd.DataFrame({"y_real":y_train["medv"], "y_prezis":y_prezis.flatten()})
real_vs_prezis_df
```

```python
real_vs_prezis_df["score"] = real_vs_prezis_df["y_real"] - real_vs_prezis_df["y_prezis"]
real_vs_prezis_df
```

```python
real_vs_prezis_df["score"] = (real_vs_prezis_df["y_real"] - real_vs_prezis_df["y_prezis"]) ** 2
real_vs_prezis_df
```

```python
real_vs_prezis_df["score"] = np.sqrt((real_vs_prezis_df["y_real"] - real_vs_prezis_df["y_prezis"]) ** 2)
real_vs_prezis_df
```

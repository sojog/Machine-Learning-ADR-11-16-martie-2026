```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("datasets/Boston_cu_headere.csv", index_col=0)
df
```

```python
# pentru a lua o coloana cu 2 seturi de paranteze drepte -> DataFrame
y = df[ [ "medv"] ]
x = df.drop("medv", axis=1) # axis=0 orizontala si axis=1 este pe verticala
```

```python
# x_scaled = StandardScaler().fit_transform(x)
# x_scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)#, random_state=142)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline_steps = [
    ("scaler", StandardScaler() ), ( "model", LinearRegression())
]
```

```python
linear_pipeline_model = Pipeline(steps=pipeline_steps)
linear_pipeline_model
```

```python
linear_pipeline_model.fit(x_train, y_train)
```

```python
linear_pipeline_model.score(x_test, y_test)
```

```python
knn_model = KNeighborsRegressor()
grid_model = GridSearchCV(estimator = knn_model, param_grid= {
    'n_neighbors':range(2, 11)
})
```

```python
knn_pipeline_steps = [
    ("scaler", StandardScaler() ), ( "model", grid_model)
]
```

```python
knn_pipeline_model = Pipeline(steps=knn_pipeline_steps)
knn_pipeline_model
```

```python
knn_pipeline_model.fit(x_train, y_train)
```

```python
knn_pipeline_model.score(x_test, y_test)
```

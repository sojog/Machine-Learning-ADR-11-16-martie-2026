```python
a, b = m
```

```python
print("a=", a)
print("b=", b)
```

```python
a, b = b, a
print("a=", a)
print("b=", b)
```

# test_size -> pentru a rezolva proportia de test vs train

```python
x = np.arange(1, 11)
y = 2 * x + 1
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5) # Valoarea recomandata este 0.2 (20%)
plt.scatter(x_train, y_train);
plt.scatter(x_test, y_test);
```

# random_state = 42 -> impartirea nu se mai face random 

```python
x = np.arange(1, 11)
y = 2 * x + 1
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # Valoarea recomandata este 0.2 (20%)
plt.scatter(x_train, y_train);
plt.scatter(x_test, y_test);
```

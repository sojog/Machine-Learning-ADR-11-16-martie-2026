```python
y = pd.DataFrame(iris_dataset.target, columns=["Category"])
y
```

```python
y["Category"].unique()
```

```python
iris_dataset.target_names
```

```python
y["Category_name"] = y["Category"].map({ index:value for index, value in enumerate(iris_dataset.target_names) })
y
```

```python
x_vs_y = pd.concat([x, y], axis=1)
x_vs_y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn_classifier_model = KNeighborsClassifier()
knn_classifier_model
```

```python
grid_search_model = GridSearchCV(knn_classifier_model, {"n_neighbors": range(2, 8)})
grid_search_model
```

```python
y_train = y_train.drop("Category_name", axis=1)
y_test = y_test.drop("Category_name", axis=1)
```

```python
grid_search_model.fit(x_train, y_train)
```

```python
grid_search_model.score(x_test, y_test)
```

# Blur image

```python
blurred_image = cv2.blur(gray_cat_image, (3, 3))
cv2.imwrite("blurred_image.jpeg", blurred_image)
```

```python
blurred_image = cv2.blur(gray_cat_image, (7, 7))
cv2.imwrite("blurred_image_v2.jpeg", blurred_image)
```

```python
top_sobel_filter = np.array([[1, 2, 1],[0, 0, 0],[-1, -2, -1]])
top_sobel_filter
```

```python
top_sobel_image = cv2.filter2D(gray_cat_image, -1,  kernel=top_sobel_filter)
cv2.imwrite("top_sobel_image.jpeg", top_sobel_image)
```

```python
plt.imshow(top_sobel_image, cmap="gray")
```

```python
left_sobel_filter = top_sobel_filter.T
left_sobel_filter
```

```python
left_sobel_image = cv2.filter2D(gray_cat_image, -1,  kernel=left_sobel_filter)
cv2.imwrite("top_sobel_image.jpeg", left_sobel_image)
plt.imshow(left_sobel_image, cmap="gray")
```

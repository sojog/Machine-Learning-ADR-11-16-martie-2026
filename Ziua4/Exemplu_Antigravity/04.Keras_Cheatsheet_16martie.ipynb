# TensorFlow Keras Cheatsheet 🚀
Acest notebook conține cele mai utilizate concepte, arhitecturi și comenzi pentru modulul `keras` din `tensorflow`.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow Version:", tf.__version__)
```

---
## 1. Crearea unui Model (Model Building)
Există două moduri principale de a construi modele în Keras: **Sequential API** (Simplu) și **Functional API** (Avansat).

### a) API-ul Secvențial (Sequential API)
Este ideal pentru arhitecturi simple, unde straturile sunt adăugate unul după altul.

```python
model_seq = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(784,)),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

# Alternativ (creare pas cu pas):
# model_seq = keras.Sequential()
# model_seq.add(layers.Dense(64, activation='relu', input_shape=(784,)))
# model_seq.add(layers.Dense(10, activation='softmax'))

model_seq.summary() # Afișează rezumatul arhitecturii
```

### b) API-ul Funcțional (Functional API)
Este folosit pentru modele mai complexe (ex: inputuri multiple, outputuri asigurate, rețele reziduale/skip connections).

```python
inputs = keras.Input(shape=(784,))
x = layers.Dense(64, activation='relu')(inputs)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(10, activation='softmax')(x)

model_func = keras.Model(inputs=inputs, outputs=outputs, name="model_functional")
model_func.summary()
```

## 3. Compilarea Modelului (Compile)
Înainte de antrenare, modelul trebuie configurat. Aici definești optimizatorul, funcția de pierdere (loss) și metricile de evaluare.

```python
model_seq.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    
    # Funcția de LOSS recomandată pe tipul problemei: 
    # 'binary_crossentropy'              -> Clasificare binară (2 clase)
    # 'categorical_crossentropy'         -> Clasificare multi-clasă (one-hot encoding)
    # 'sparse_categorical_crossentropy'  -> Clasificare multi-clasă (label-uri intregi ca 0, 1, 2)
    # 'mse' (Mean Squared Error)         -> Regresie
    loss='sparse_categorical_crossentropy',
    
    metrics=['accuracy']
)
```

## 5. Callbacks (Control Avansat al Antrenamentului)
Sunt funcții pe care Keras le apelează în diferite etape ale antrenării (ex: la sfârșitul fiecărei epoci). Sunt esențiale pentru Early Stopping, reducerea learning rate-ului sau salvarea celui mai bun model (Checkpoints).

```python
callbacks = [
    # Oprește antrenamentul dacă 'val_loss' nu se îmbunătățește timp de 3 epoci
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True
    ),
    
    # Salvează doar modelul când are cea mai bună performanță
    keras.callbacks.ModelCheckpoint(
        filepath='best_model.keras',
        monitor='val_loss',
        save_best_only=True
    ),
]

# Cum se atașează de .fit()
# history = model_seq.fit(x_train, y_train, epochs=20, callbacks=callbacks)
```

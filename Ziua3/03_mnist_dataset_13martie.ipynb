# Crearea modelului

```python
model = tf.keras.models.Sequential(
    [
        ## INPUT LAYER - va fi intotdeauna in functie de tipul datelor de intrare
          tf.keras.layers.Flatten(input_shape=(28, 28)),

        ## Numarul de straturi, numarul de neuroni per strat, functia de activare pot varia
         tf.keras.layers.Dense(10, activation="relu"),




        ## OUTPUT Layer - va fi intotdeauna la fel in functie de output datelor (ce trebuie sa intoarca)
          tf.keras.layers.Dense(10, activation="softmax")

    ]
)
model.summary()
```

```python
7840 + 10
```

# sparse_categorical_crossentropy -> pentru mai multe categorii
# binary_crossentropy -> pentru 2 categorii

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test, y_test)
```

```python
x_train_scaled = x_train / 255
x_test_scaled = x_test / 255
```

```python
model2 = tf.keras.models.Sequential(
    [
        ## INPUT LAYER - va fi intotdeauna in functie de tipul datelor de intrare
          tf.keras.layers.Flatten(input_shape=(28, 28)),

        ## Numarul de straturi, numarul de neuroni per strat, functia de activare pot varia
         tf.keras.layers.Dense(100, activation="relu"),



        ## OUTPUT Layer - va fi intotdeauna la fel in functie de output datelor (ce trebuie sa intoarca)
          tf.keras.layers.Dense(10, activation="softmax")

    ]
)
model2.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model2.fit(x_train_scaled, y_train, epochs=10)
```

```python
loss = model2.history.history["loss"]
```

```python
plt.plot(loss)
plt.title("Loss")
plt.legend(["Training Data"])
plt.xlabel("Epoci")
```

```python
accuracy = model2.history.history["accuracy"]
plt.plot(accuracy)
plt.title("Acuratetea ")
plt.xlabel("Epoci")
plt.legend(["Training Data"])
```

```python
model4 = tf.keras.models.Sequential(
    [
        ## INPUT LAYER - va fi intotdeauna in functie de tipul datelor de intrare
          tf.keras.layers.Flatten(input_shape=(28, 28)),

        ## Numarul de straturi, numarul de neuroni per strat, functia de activare pot varia
         tf.keras.layers.Dense(100, activation="relu"),



        ## OUTPUT Layer - va fi intotdeauna la fel in functie de output datelor (ce trebuie sa intoarca)
          tf.keras.layers.Dense(10, activation="softmax")

    ]
)
model4.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model4.fit(x_train_scaled, y_train, epochs=20, validation_split=0.2)
```

```python
loss = model4.history.history["loss"]
val_loss = model4.history.history["val_loss"]
plt.plot(loss)
plt.plot(val_loss)
plt.title("Loss")
plt.legend(["Training Data", "Validation Data"])
plt.xlabel("Epoci")
```

```python
accuracy = model4.history.history["accuracy"]
val_accuracy = model4.history.history["val_accuracy"]
plt.plot(accuracy)
plt.plot(val_accuracy)
plt.title("accuracy")
plt.legend(["Training Data", "Validation Data"])
plt.xlabel("Epoci")
```
